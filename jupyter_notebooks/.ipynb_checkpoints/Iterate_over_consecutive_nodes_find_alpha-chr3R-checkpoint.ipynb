# 3. Read Original Graph in networkx, draw adjacency

```python
def read_networks_as_graph(path, if_DNA_string=True, node_range=None):
    if if_DNA_string:
        cur_df = pd.read_csv(path, index_col = 0)
        edgelist = list(cur_df.apply(lambda x: (x.edge_list.split(',')), axis = 1).values)
#         edgelist = np.genfromtxt(path, delimiter=',', dtype=str, skip_header=1)
#         edgelist = edgelist[:, 1:]
#         edgelist = edgelist.tolist()
        if 'ChIA_Drop' not in path:
            ### Delete unwanted characters
            for edge in edgelist:
                edge[0] = edge[0][3:-1]
                edge[1] = edge[1][2:-3]
    else:
        edgelist = np.genfromtxt(path, delimiter=',', dtype=str)
        edgelist = edgelist.tolist()

    G = nx.Graph()
    G.add_edges_from(edgelist)

    print('number of nodes=', len(G.nodes))
    print('number of edges=', len(G.edges))
    return G
```

## reorder dictionaries according to their importance score

```python
def get_importance_from_At(A_t):
    diag = A_t.diagonal()
    sum_diag = sum(diag)
    score = diag / sum_diag
    return score
```

```python
importance_score = get_importance_from_At(A_t)
subtitles = [f'online cvxNDL\n dictionaries {x}\n score {importance_score[x]:.2f}' for x in range(ocmf_dict_val_df.shape[0])]
```

## Get alpha for each block

```python
def get_alpha_of_row(row, D_hat):
    x_sample = row.values
    lars_lasso = Lasso(alpha = 0, max_iter = 500)
    lars_lasso.fit(D_hat.T, x_sample)
    
    alpha = lars_lasso.coef_
    
#     x_hat = D_hat.T @ alpha
#     return x_hat
    return alpha
```

```python
D_hat = sorted_ocmf_dict_val_df_by_At.values

learned_alphas = np.vstack(test_df.progress_apply(lambda row: get_alpha_of_row(row, D_hat), axis = 1).values)
```

```python
alpha_column = [f'alpha_{i}' for i in descending_order_of_At]
df_alpha = pd.DataFrame(learned_alphas, columns = alpha_column)
```

```python
# row_max_alpha = lambda row: alpha_column[row.abs().argmax()]
# df_alpha['max_alpha'] = df_alpha.progress_apply(row_max_alpha, axis = 1)

def find_max_second_max_and_difference(row):
    sorted_alphas = np.sort(row.abs())
    max_alpha = sorted_alphas[-1]
    second_max_alpha = sorted_alphas[-2]
    diff = max_alpha - second_max_alpha
    
    return diff

threshold_of_difference = 0.05

df_alpha_selected_rows = df_alpha[df_alpha.progress_apply(find_max_second_max_and_difference, axis = 1) > threshold_of_difference]

# df_alpha.progress_apply(find_max_second_max_and_difference, axis = 1).hist()
```

# Select interesting clusters, draw Adj matrix and dictionary.

```python
def plot_one_row(ax, row, k = 21, threshold = 1, title = ''):
    row_np = np.array(row)
    if threshold is not None:
        bin_row = (row_np >= threshold).astype(int)
        bin_row_mat = bin_row.reshape(k, k)
    else:
        bin_row_mat = row_np.reshape(k, k)
    
    ax.imshow(bin_row_mat, cmap = 'Greys')
    ax.set_title(title)
    ax.set_xticks([])
    ax.set_yticks([])
    return bin_row_mat, ax
```

```python
def plot_df_in_grid(df_in, grid_size = (2, 5), figsize = (10, 4), 
                    k = 21, threshold = 1, title_prefix = 'row', title = '', 
                   sub_titles = None, p2savefig = None):
    fig = plt.figure(figsize = figsize)
    axes = fig.subplots((*grid_size))
    
    for row_idx in range(grid_size[0]):
        for col_idx in range(grid_size[1]):
            ax = axes[row_idx][col_idx]
            
            idx = row_idx * grid_size[1] + col_idx
            if idx < df_in.shape[0]:
                cur_row = df_in.iloc[idx]
            else:
                cur_row = np.zeros_like(df_in.iloc[0].values)
            
            if sub_titles is None:
                subtitle =  f'{title_prefix}_{idx}'
            else:
                if idx < len(sub_titles):
                    subtitle = sub_titles[idx]
                else:
                    subtitle = ''
                
            plot_one_row(ax, cur_row, k = k, threshold= threshold, title = subtitle)
    plt.suptitle(title, y = 1.05)
    plt.tight_layout()
    
    if p2savefig is None:
        plt.show()    
        plt.close()
    elif p2savefig == 'combined':
        return fig
    else:
        plt.savefig(p2savefig)
        plt.close()
```

```python
def draw_arc_of_graph(H_directed, pos, ax = None, title = '', p2savefig = None):
    if ax is None:
        fig, ax = plt.subplots(figsize = (20, 3))

    nx.draw_networkx_nodes(H_directed, pos, 
                           node_size = 100, 
                           ax = ax)
    nx.draw_networkx_edges(H_directed, pos, 
                           connectionstyle='arc3,rad=-0.3', 
                           arrowsize= 0.1,
                           ax = ax)

    # skip some value of nodes that are close to each other.
    old_x_value = -100
    for value, loc in pos.items():
        x, y = loc 
        if (x - old_x_value) < 3:
            continue
        old_x_value = x
        ax.text(x, y - 20, value,
                rotation = 315)
    
    plt.title(title)
    plt.ylim(-20, 100)
    
    if p2savefig is None:    
        plt.show()
        plt.close()
    elif p2savefig == 'combined':
        return fig
    else:
        plt.savefig(p2savefig)
        plt.close()
    return ax
```

```python
# list_of_cluster = [19, 20, 4]

for cluster in descending_order_of_At:
    print('-' * 7, f'start of dicitonary {cluster}', '-' * 7)
    cur_dictionary = ocmf_dict_val_df.loc[cluster]
    
    # This one will select the row that has alpha_{cluster} being largest.
#     selected_index = (df_alpha.max_alpha == f'alpha_{cluster}')
    
    # This one will select alpha_{cluster} first, sort by the column's abs value, and select the top rows.
    df_sorted_alpha = df_alpha_selected_rows.sort_values(by = f'alpha_{cluster}', ascending = False, key = abs)
    
    # check if the current alpha is the largest for the current cluster/dictionary.
    matched_max_alpha = df_sorted_alpha[f'alpha_{cluster}'] == df_sorted_alpha.max(axis = 1)
    df_sorted_alpha = df_sorted_alpha[matched_max_alpha]
    
    selected_index = df_sorted_alpha.index
    
    if sum(selected_index) == 0:
        continue
    
        
    selected_df = test_df.loc[selected_index].head(10)
    selected_nodes = test_node_df.loc[selected_df.index]
    
#     if len(selected_df) >= 10:
#         selected_df = selected_df.sample(10, replace = False)
#         selected_nodes = selected_nodes.loc[selected_df.index]
    
    start_idx = selected_nodes.node_0.values
    subtitles = [f'{x} - V{int(x[1:]) + 21}' for x in start_idx]
    
    
#     plot_df_in_grid(selected_df, sub_titles=subtitles)
    
    # For each selected subgraph, draw the arc plot of nodes.
    for idx, cur_adj in selected_df.iterrows():
        if (cur_adj == 0).all():
            continue
            
        cur_nodes = selected_nodes.loc[idx].values
        cur_adj = selected_df.loc[idx].values

        # Plot the dictionary and current region side by side. 
        fig_dict = plt.figure()
        ax_dict = fig_dict.add_subplot(121)
        
        plot_one_row(ax_dict, ocmf_dict_val_df.loc[cluster], title = f'dictionary {cluster}', threshold = None)
#     plt.show()
#     plt.close()
        ax_region = fig_dict.add_subplot(122)
        plot_one_row(ax_region, cur_adj, title = f'region: {cur_nodes[0]} - {cur_nodes[-1]}')
               
        # create networkx graph from adj and node id.
        k = len(cur_nodes)
        tmp_df = pd.DataFrame(data = cur_adj.reshape(k, k), index = cur_nodes, columns=cur_nodes)
        tmp_G = nx.from_pandas_adjacency(tmp_df)
        
        # compute position for each node.
        node_pos = np.array([int(x[1:]) for x in cur_nodes])
        min_node_pos = min(node_pos)
        max_node_pos = max(node_pos)
        node_pos_span = max_node_pos - min_node_pos
        pos = {cur_nodes[i] : ((node_pos[i] - min_node_pos) / node_pos_span * 100, 0) \
               for i in range(len(cur_nodes))}
        
        # create diGraph for plotting.
        cur_G = nx.DiGraph()
        cur_G.add_nodes_from(tmp_G.nodes(data = True))
        cur_G.add_edges_from(tmp_G.edges(data = True))
        
        cur_alpha = df_sorted_alpha.loc[idx][f'alpha_{cluster}']
        cur_title = f'{cur_nodes[0]} - {cur_nodes[-1]}, alpha = {cur_alpha:.2f}'
        
        p2saved = f'/data/shared/jianhao/online_cvxNDL_results/regions_dominated_by_one_dictionary/{chromo}/region_edge_list/dictionary_{cluster}'
        if not osp.isdir(p2saved):
            os.makedirs(p2saved)
            
        p2edge_list = osp.join(p2saved, f'edge_list_{cur_nodes[0]}_to_{cur_nodes[-1]}.txt')
        with open(p2edge_list, 'w') as f:
            for line in list(cur_G.edges()):
                f.write(','.join(line))
                f.write('\n')
        

        p2img = osp.join(p2saved, f'arc_plot_regions_use_mostly_dicitonary_{cluster}_{cur_nodes[0]}_to_{cur_nodes[-1]}')
        
#         draw_arc_of_graph(cur_G, pos, title = cur_title, p2savefig=p2img)
        fig_arc = draw_arc_of_graph(cur_G, pos, title = cur_title, p2savefig='combined')
    
        combined_dict_and_regions = PdfPages(p2img + '.pdf')
        combined_dict_and_regions.savefig(fig_dict)
        combined_dict_and_regions.savefig(fig_arc)
        combined_dict_and_regions.close()
        plt.close()
    
        
        
    print('-' * 7, f'end of dicitonary {cluster}', '-' * 7)
```
